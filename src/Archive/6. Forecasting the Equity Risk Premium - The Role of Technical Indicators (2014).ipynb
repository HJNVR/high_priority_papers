```python
#import libraries
import numpy as np 
import pandas as pd 
import yfinance as yf
import datetime
import pandas_datareader
```

```python
#define start and end date
start_date="1999-12-31"
end_date="2021-01-01"
```

```python
#import list of S&P500 company tickers
stocks = pd.read_csv('S&P500 companies list (2000 to 2020).csv')
tickers= stocks['ticker'].values
```

```python
#function for stock data download from yahoo finance
def download_stock(ticker):
    df_ticker = yf.download(ticker, start=start_date, end=end_date, interval='1mo')
    return df_ticker
```

```python
#create dictionary with key as stock ticker and value as dataframe downloaded from yahoo finance
dic = {}
for t in tickers:
    dic[t] = download_stock(t)
```

```python
#functions to calculate stock specific features

#Calculate MA
def calculate_MA(df):
    df['MA1'] = df['Close'].rolling(1, min_periods=1).mean()
    df['MA2'] = df['Close'].rolling(2, min_periods=1).mean()
    df['MA3'] = df['Close'].rolling(3, min_periods=1).mean()
    df['MA9'] = df['Close'].rolling(9, min_periods=1).mean()
    df['MA12'] = df['Close'].rolling(12, min_periods=1).mean()
    return df

#Calculate Momentum
def calculate_momentum(df):
    df['Close_l1'] = df['Close'].shift(1)
    df['Close_l9'] = df['Close'].shift(9)
    df['Close_l12'] = df['Close'].shift(12)
    return df

#Calculate D
def calculate_D(df):
    D = []
    D.append(1)
    for i in range(1,len(df['Close'])):
        if df['Close'][i] >= df['Close_l1'][i]:
            D.append(1)
        else:
            D.append(-1)
    df['D'] = D
    return df

#calculate moving average of OBV
def calculate_MAOBV(df):
    OBV=[]
    for i in range(len(df['Close'])):
        OBV.append(df['Volume'][i]*df['D'][i])
    df['OBV'] = OBV
    df['MAOBV1'] = df['OBV'].rolling(1, min_periods=1).mean()
    df['MAOBV2'] = df['OBV'].rolling(2, min_periods=1).mean()
    df['MAOBV3'] = df['OBV'].rolling(3, min_periods=1).mean()
    df['MAOBV9'] = df['OBV'].rolling(9, min_periods=1).mean()
    df['MAOBV12'] = df['OBV'].rolling(12, min_periods=1).mean()
    return df

#Calculate features
#1 MA(1,9)
def MA_1_9(df):
    MA_1_9 = []
    for i in range(len(df['Close'])):
        if df['MA1'][i] >= df['MA9'][i]:
            MA_1_9.append(1)
        else:
            MA_1_9.append(0)
    df['MA_1_9'] = MA_1_9
    return df

#2 MA(1,12)
def MA_1_12(df):
    MA_1_12 = []
    for i in range(len(df['Close'])):
        if df['MA1'][i] >= df['MA12'][i]:
            MA_1_12.append(1)
        else:
            MA_1_12.append(0)
    df['MA_1_12'] = MA_1_12
    return df

#3 MA(2,9)
def MA_2_9(df):
    MA_2_9 = []
    for i in range(len(df['Close'])):
        if df['MA2'][i] >= df['MA9'][i]:
            MA_2_9.append(1)
        else:
            MA_2_9.append(0)
    df['MA_2_9'] = MA_2_9
    return df

#4 MA(2,12)
def MA_2_12(df):
    MA_2_12 = []
    for i in range(len(df['Close'])):
        if df['MA2'][i] >= df['MA12'][i]:
            MA_2_12.append(1)
        else:
            MA_2_12.append(0)
    df['MA_2_12'] = MA_2_12
    return df

#5 MA(3,9)
def MA_3_9(df):
    MA_3_9 = []
    for i in range(len(df['Close'])):
        if df['MA3'][i] >= df['MA9'][i]:
            MA_3_9.append(1)
        else:
            MA_3_9.append(0)
    df['MA_3_9'] = MA_3_9
    return df

#6 MA(3,12)
def MA_3_12(df):
    MA_3_12 = []
    for i in range(len(df['Close'])):
        if df['MA3'][i] >= df['MA12'][i]:
            MA_3_12.append(1)
        else:
            MA_3_12.append(0)
    df['MA_3_12'] = MA_3_12
    return df

#7 MOM9
def MOM9(df):
    MOM9 = []
    for i in range(len(df['Close'])):
        if df['Close'][i] >= df['Close_l9'][i]:
            MOM9.append(1)
        else:
            MOM9.append(0)
    df['MOM9'] = MOM9
    return df

#8 MOM12
def MOM12(df):
    MOM12 = []
    for i in range(len(df['Close'])):
        if df['Close'][i] >= df['Close_l12'][i]:
            MOM12.append(1)
        else:
            MOM12.append(0)
    df['MOM12'] = MOM12
    return df

#9 VOL(1,9)
def VOL_1_9(df):
    VOL_1_9 = []
    for i in range(len(df['Close'])):
        if df['MAOBV1'][i] >= df['MAOBV9'][i]:
            VOL_1_9.append(1)
        else:
            VOL_1_9.append(0)
    df['VOL_1_9'] = VOL_1_9
    return df

#10 VOL(1,12)
def VOL_1_12(df):
    VOL_1_12 = []
    for i in range(len(df['Close'])):
        if df['MAOBV1'][i] >= df['MAOBV12'][i]:
            VOL_1_12.append(1)
        else:
            VOL_1_12.append(0)
    df['VOL_1_12'] = VOL_1_12
    return df

#11 VOL(2,9)
def VOL_2_9(df):
    VOL_2_9 = []
    for i in range(len(df['Close'])):
        if df['MAOBV2'][i] >= df['MAOBV9'][i]:
            VOL_2_9.append(1)
        else:
            VOL_2_9.append(0)
    df['VOL_2_9'] = VOL_2_9
    return df

#12 VOL(2,12)
def VOL_2_12(df):
    VOL_2_12 = []
    for i in range(len(df['Close'])):
        if df['MAOBV2'][i] >= df['MAOBV12'][i]:
            VOL_2_12.append(1)
        else:
            VOL_2_12.append(0)
    df['VOL_2_12'] = VOL_2_12
    return df

#13 VOL(3,9)
def VOL_3_9(df):
    VOL_3_9 = []
    for i in range(len(df['Close'])):
        if df['MAOBV3'][i] >= df['MAOBV9'][i]:
            VOL_3_9.append(1)
        else:
            VOL_3_9.append(0)
    df['VOL_3_9'] = VOL_3_9
    return df

#14 VOL(3,12)
def VOL_3_12(df):
    VOL_3_12 = []
    for i in range(len(df['Close'])):
        if df['MAOBV3'][i] >= df['MAOBV12'][i]:
            VOL_3_12.append(1)
        else:
            VOL_3_12.append(0)
    df['VOL_3_12'] = VOL_3_12
    return df

#clean-up unnecessary columns
def cleanup(df):
    return df.drop(['Open','High','Low','Adj Close','Close','Volume',
                    'MA1','MA2','MA3','MA9','MA12',
                    'Close_l1','Close_l9','Close_l12',
                   'D','OBV','MAOBV1','MAOBV2','MAOBV3','MAOBV9','MAOBV12'],axis=1)
```

```python
#function to calculate all features
def cal_export(df):
    df = calculate_MA(df)
    df = calculate_momentum(df)
    df = calculate_D(df)
    df = calculate_MAOBV(df)
    df = MA_1_9(df)
    df = MA_2_9(df)
    df = MA_3_9(df)
    df = MA_1_12(df)
    df = MA_2_12(df)
    df = MA_3_12(df)
    df = MOM9(df)
    df = MOM12(df)
    df = VOL_1_9(df)
    df = VOL_2_9(df)
    df = VOL_3_9(df)
    df = VOL_1_12(df)
    df = VOL_2_12(df)
    df = VOL_3_12(df)
    df = cleanup(df)
    return df
```

```python
#calculate features and export as csv file
for k,v in dic.items():
    if v.shape[0] > 12:
        cal_export(v).to_csv(k+'.csv')
```
