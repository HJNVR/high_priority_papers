# Investing.com

```python
import investpy
```

```python
start_date = "01/01/2000"
end_date = "31/12/2021"
```

```python
#indices: stocks, funds, ETFs, currency crosses, certificates, bonds, commodities, indices, and cryptos
search_result = investpy.search_quotes(text='Brent', products=['commodities'], n_results=1)
print(search_result)
```

```python
OIL_BRENT = investpy.get_commodity_historical_data(commodity='Brent Oil', from_date=start_date, to_date=end_date)
OIL_BRENT['Relative Change'] = (OIL_BRENT['Close'] - OIL_BRENT['Close'].shift()) / OIL_BRENT['Close'].shift()
OIL_BRENT
```
